```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.ticker as ticker
from matplotlib.lines import Line2D


dataset = 'Planar'
print(dataset)

uniform = np.array([0, 0.05, 0.12, 0.23, 0.77, 0.44])
marginal = np.array([0, 0.27, 0.63, 0.71, 0.77, 0.84])
absorbing = np.array([0.11, 0.56, 0.65, 0.53, 0.44, 0.44])
masking = np.array([0, 0, 0, 0, 0, 0])

fig, ax1 = plt.subplots(figsize=(4., 2.))
num_steps = np.array(['0k', '20k', '40k', '60k', '80k', '100k'])

colors = sns.color_palette("mako", 12)

ax1.tick_params(axis='y')
ax1.set_ylabel('V.U.N')
ax1.plot(num_steps, marginal, label='Marginal')
ax1.plot(num_steps, uniform, label='Uniform')
ax1.plot(num_steps, masking, label='Masking')
ax1.plot(num_steps, absorbing, label='Absorbing')

ax1.set_xlabel(r'\# Steps')
ax1.set_ylabel('V.U.N. (Planar)')
ax1.grid(True, alpha=0.75); ax1.xaxis.grid(False)
ax1.yaxis.set_major_locator(ticker.MaxNLocator(3))

# Set larger font sizes for the axes
for item in ([ax1.xaxis.label, ax1.yaxis.label]):
    item.set_fontsize(14)
for item in (ax1.get_xticklabels() + ax1.get_yticklabels()):
    item.set_fontsize(14)
ax1.title.set_fontsize(14)

fig.legend(fontsize=12, loc='upper left', bbox_to_anchor=(0.9, 0.8))

plt.savefig(f'./{dataset}_num_steps_no_label.pdf', bbox_inches='tight')
plt.show()
plt.show()
```
